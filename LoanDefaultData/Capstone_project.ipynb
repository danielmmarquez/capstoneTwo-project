## Introduction

 
The purpose of this data science project is to have a better selection process for individuals who apply for a loan. overall. there are 4 main outcomes for a finance company when loaning a customer money
* 1: customer manages to continuously make one time payments for the entirety of the loan agreement which is the most profitable outcome. 
* 2: the loan company grants a loan to a customer and they default on their loan meaning that the loan company is at a loss on that contract
* 3: the loan company denies the loan application when the customer would not have defaulted. the loan company missed out on potential profit.
* 4: The loan company denies a customer, the customer would have defaulted if approved. the loan company made a good decision denying the application
The goal is to maximize profits and minimize losses meaning, all applicants that will not default get approved, all others get denied.

# 2. method

Given that the purpose of this project is to determine which applicants will default and which will not default, we need to choose a classification model and work with the model that yields the best performance. in this case, three models where used
* KNN classifier
* XGBoost classification
* Random Forest Classifier
Ultimately, KNN had the best performance.


# 3. Data Cleaning

As any data related project, you want to make sure that there are no missing values. since there are two data sets, both must be cleared of any null or missing values. Once the missing values are taken care of, merging the data frames on the "TARGET" column is done in order to have one unified dataframe. The "TARGET" column indicates if a customer is having difficulty paying their loans so this is a key column that is used to train the model.

# 5 Modeling 

KNN performed the best; however, a specific k-value must be chosen. Grid search was first implemented and returned a k-value of 11 would be the optimal value to move forward with. just to verify the error rate v.s. k-value was plotted. as we can see, as k increases the error rate tends to rise as well as shown below.

now plotting the accuracy vs k-value. As K increases, the accuracy decreases. Notice however, K’s 1 through 20 still maintain an accuracy above 60%.

Recall than the data being used is unbalanced and contains mostly approved applications. SMOTE was used in order to create a well balanced model. Before implementing SMOTE, 20% of the original unbalanced set was saved for testing purposes. Ultimately, KNN with 11 neighbors ended up performing the best with the 20% imbalanced data that was saved for testing. the overall results are listed below<br>
