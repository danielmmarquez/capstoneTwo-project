# Capstone Two: Preprocessing and Training Data Development

In the EDA portion of this investigation, I concluded that the best approach would be to issolate the status of the contracts by either approved loans or refused loans which the data i will now be working with. 

### 1. Sourcing and loading

#### 1a. Import relevant libraries 

```python
#importing nessicaty packages
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

#### 1b. Load the data

```python
#importing the data frames created in the EDA notebook where the approved and refused contractes
#where seperated into two different DF and any missing data was dropped :)
dropped_na_approved = pd.read_csv('dropped_na_approved.csv')
dropped_na_refused = pd.read_csv('dropped_na_refused.csv')
#dropping the column 'unnamed: 0' because it has no relivance 
dropped_na_approved.drop('Unnamed: 0', axis = 1, inplace = True)
dropped_na_refused.drop('Unnamed: 0', axis = 1, inplace = True)
```

### 1c. checking for null values, minor data cleaning

```python
print(dropped_na_approved.isna().sum())
print(dropped_na_refused.isna().sum())
```

there is some missing data kept as 'XNA' which still remains after EDA so we will simply remove them from the data frames

```python
refused_df = dropped_na_refused[dropped_na_refused['NAME_YIELD_GROUP'] != 'XNA'] 
approved_df = dropped_na_approved[dropped_na_approved['NAME_YIELD_GROUP'] != 'XNA'] 
```

### 2. Creating dummy vareables

Creating dummy variables for the object dtypes and bringing it all together

```python
dfo_approved = approved_df.select_dtypes(include=['object']) #select object type columns
approved_df = pd.concat([approved_df.drop(dfo_approved, axis=1), pd.get_dummies(dfo_approved)], axis=1)
```

```python
dfo_refused = refused_df.select_dtypes(include=['object']) #select object type columns
refused_df = pd.concat([refused_df.drop(dfo_refused, axis=1), pd.get_dummies(dfo_refused)], axis=1)
```

Making sure they where murged together successfully

### 3. Standardizing numerical data

```python
scaler = StandardScaler() #initiallizing StandardScaler

#scaling the approved dataframe
scaled_approved = scaler.fit_transform(approved_df) #standardizing the approved data
                #turning the array into a dataframe again 
scaled_approved = pd.DataFrame(scaled_approved, columns = approved_df.columns)

#scaling the refused dataframe
scaled_refused = scaler.fit_transform(refused_df) #standardizing the refused data
                #turning the array into a datafreame again 
scaled_refused = pd.DataFrame(scaled_refused, columns = refused_df.columns)
```

```python
#verifying that the dataframe was scaled correctly 
display(scaled_refused, scaled_approved)
```

### 4. splitting data into training and testing sets

```python
x = scaled_approved['AMT_APPLICATION'] #choosing x-values
y = scaled_approved['AMT_APPLICATION']#choosing y-values
#the 'a' indicates these x's and y's are for the approved DF
Xa_train, Xa_test, ya_train, ya_test = train_test_split(x, y, test_size = 0.25, random_state = 125)
```

we have now split the data into 75% teaining set and 25% testing set for the approved loans. we just need to do the same for the refused loans

```python
x = scaled_refused['AMT_APPLICATION'] #choosing x-values
y = scaled_refused['AMT_APPLICATION']#choosing y-values
Xr_train, Xr_test, yr_train, yr_test = train_test_split(x, y, test_size = 0.25, random_state = 125)
```
