## introduction

In this notebook, data containing information of defaulted loans is gathered, organized, and will ensure it's well defined. we will not drop any data in this portion of analysis however, we will decide which of the two data sets will be useful in our exploration. 
The purpose of this data science project is to have a better selection process for individuals who apply for a loan. over all. there are 4 main outcomes for a finance company when loaning a customer money
* 1: customer manages to continuously make onetime payments for the entirety of the loan agreement which is the most profitable outcome. 
* 2: the loan company grants a loan to a customer and they default on their loan meaning that the loan company is at a loss on that contract
* 3: the loan company denies the loan application when the customer would not have defaulted. the loan company missed out on potential profit.
* 4: The loan company denies a customer, the customer would have defaulted if approved. the loan company made a good decision denying the application
The goal is to maximize profits and minimize losses meaning, all applicants that will not default get approved, all others get denied.

As we can see, the column description file does indeed tell us what the column names are for the other two files however, it would make more sence if we transpose the data in order to see all columns. we also have no need for the first column labeled as 'Unnamed: 0' so we can drop it before transposing the DF
